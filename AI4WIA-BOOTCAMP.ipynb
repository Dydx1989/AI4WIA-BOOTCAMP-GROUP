# *Step 1: Importing the dataset*
# *Step 2: Data pre-processing*
# *Step 3: Splitting the test and train sets*
# *Step 4: Fitting the linear regression model to the training set*
# *Step 5: Predicting test results*
# *Step 6: Visualizing the test results*

# **1. Importing the dataset**
# *We will begin with importing the dataset using pandas and also import other libraries such as numpy and matplotlib*

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
 
dataset = pd.read_csv('crop_data.csv')
dataset.head()
```

The dataset.head() shows the first few columns of our dataset. The output of the above snippet is as follows:

# **2. Data Preprocessing**
# *Now that we have imported the dataset, we will perform data preprocessing.*

```python
X = dataset.iloc[:,:-1].values  #independent variable array
y = dataset.iloc[:,1].values  #dependent variable vector
```

# *The X is independent variable array and y is the dependent variable vector. Note the difference between the array and vector. The dependent variable must be in vector and independent variable must be an array itself.*

# **3. Splitting the dataset**
# *We need to split our dataset into the test and train set. Generally, we follow the 20-80 policy or the 30-70 policy respectively.*



```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=1/3,random_state=0)
```

# **4. Fitting linear regression model into the training set**
# *From sklearns linear model library, import linear regression class. Create an object for a linear regression class called regressor.*
# *To fit the regressor into the training set, we will call the fit method  function to fit the regressor into the training set.*
# *We need to fit X_train (training data of matrix of features) into the target values y_train. Thus the model learns the correlation and learns how to predict the dependent variables based on the independent variable.*

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train,y_train) #actually produces the linear eqn for the data
# Display the coefficients and intercept
print("Coefficients:", regressor.coef_)
print("Intercept:", regressor.intercept_)
```

```python
import statsmodels.api as sm

# Add a constant to your feature matrix for the intercept
X_train = sm.add_constant(X_train)

# Create a linear regression model
model = sm.OLS(y_train, X_train)

# Fit the model
results = model.fit()

# Display coefficients and p-values
print("Coefficients:")
print(results.params)

# Display p-values
print("\nP-values:")
print(results.pvalues)
```

# **5. Predicting the test set results**
# *We create a vector containing all the predictions of the test set salaries. The predicted salaries are then put into the vector called y_pred.(contains prediction for all observations in the test set)*

```python
y_pred = regressor.predict(X_test) 
y_pred
```

# **Plotting the points (observations)**

```python
import matplotlib.pyplot as plt

# Assuming you have X_train and y_train from your linear regression example
# Create a scatter plot of the data points
plt.scatter(X_train[:, 0], y_train, label="Data Points")

# Add the regression line
plt.plot(X_train[:, 0], regressor.predict(X_train[:, 0]), color='red', label="Regression Line")

# Add labels and a legend
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()

# Show the plot
plt.show()
```
